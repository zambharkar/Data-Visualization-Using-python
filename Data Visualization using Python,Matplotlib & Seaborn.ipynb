# Line Chart

Line charts are one of the simplest and most widely used data visualization techniques


```python
yield_apples=[0.895, 0.91, 0.919, 0.926, 0.929, 0.931]
```

we can visualize how the yield of apples changes over tme using a line chart. To draw a line chart we can use the plt.plot

```python
plt.plot(yield_apples);
```

lets enhance this plot step by step to make it more informative and beautiful

# Customizing the X-axis

The X-axis of the plot currently shows list element indexs 0 to 5. The plot would be more informative if we could show the year for which the data is being plotted.We can do this by two arguments plt.plot

```python
years = [2010, 2011, 2012, 2013, 2014, 2015]
yield_apples = [0.895, 0.91, 0.919, 0.926, 0.929, 0.931]
plt.plot(years, yield_apples);
```

# Axis Labels

we can add labels to the axes to show what each axis represent using the plt.xlabel and plt.ylabel methods

```python
plt.plot(years,yield_apples)
plt.xlabel('Year')
plt.ylabel('Yield(tons per hectare)');
```

# Plotting Multiples Lines

Its really easy to plot multiple line in same graoh.Just invoke the plt.plot function multiple times. Lets compare the yield of apples vs oranges in Kanto

```python
years=range(2000,2012)
apples=[0.895, 0.91, 0.919, 0.926, 0.929, 0.931, 0.934, 0.936, 0.937, 0.9375, 0.938, 0.9385]
oranges=[0.962, 0.941, 0.930, 0.923, 0.918, 0.908, 0.907, 0.904, 0.901, 0.898, 0.888, 0.875]
plt.plot(years,apples)
plt.plot(years,oranges)
plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)');
```

```python
plt.plot(years,apples,marker='o')
plt.plot(years,oranges,marker='x')
plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')
plt.title("crop Yields in Kanto")
plt.legend(['Apples', 'Pranges']);
```

# Styling lines and markers

```python
plt.plot(years,apples,marker='s', c='b', ls='-', lw=2, ms=8, mew=2,mec='navy')
plt.plot(years,oranges,marker='o', c='r', ls='--', lw=3, ms=10, alpha=.5 )
plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')
plt.title("Crop Yields in Kanto")
plt.legend(['Apples', 'oranges']);
```

# Scatter Plot

in a scatter plot the values of 2 variables are plotted as points on a 2-dimensional grid.Additionally you can also use a third variable to determine the size or color of the point s lets try out an example

```python
flowers_df=sns.load_dataset('iris')
flowers_df
```

```python
flowers_df.species.unique()
```

lets try to visualization the retaion between sepal lenght and sepal widht our first instinct might be to create a line chart using plt.plot however the output is not very informative as thre are too many combinations of the two properties within the dataset and there dosent seem to be simple relationship between them

```python
flowers_df
```

```python
plt.plot(flowers_df.sepal_length, flowers_df.sepal_width);
```

```python
sns.scatterplot(flowers_df.sepal_length,flowers_df.sepal_width);
```

# Adding Hues

Notice how the points in the above plot seem to form distinct clusters with some outliers we can color the dots using flowers species as a hue we can also make the point larger using the s argument

```python
sns.scatterplot(flowers_df.sepal_length, flowers_df.sepal_width,hue=flowers_df.species, s=100);
```

# Customizing seaborn Figures

Since seaborn uses matplotlib plotting functions internally we can use functions like plt.figure and plt.title to modify the figure

```python
plt.figure(figsize=(12,6))
plt.title('sepal Dimensions')
sns.scatterplot(flowers_df.sepal_length, flowers_df.sepal_width,hue=flowers_df.species, s=100);
```

# Plotting Using pandas data Frames


seaborn has inbuilt support for pandas data frames instead of passing each column as a series you can also pass column names and the data frame

```python
plt.title('sepal Dimesions')
sns.scatterplot('sepal_length', 'sepal_width', hue='species', s=100, data=flowers_df);
```

# Histogram


A histogram represent the distribution of data by forming bins along the range of the data and the drawing bars to show the number of observation that fall in each bin
we can use the plt.hist function to create a histogram

```python
#Load data into a pandas dataframe
flowers_df=sns.load_dataset('iris')
```

```python
flowers_df.sepal_width
```

```python
flowers_df.describe()
```

```python
plt.title('Distribution of sepal width')
plt.hist(flowers_df.sepal_width);
```

we can immediately see that vlues of sepal width fall in the range 2.0-4.5 and around 35 values are in the range 2.9-3.1 which seems to be the laegest bin


# Controlling the size and number of bins

we can control the number of bins or the size of each bin using the bins argument

```python
#Specifying the number of bins
plt.hist(flowers_df.sepal_width, bins=5);
```

```python
import numpy as np
#Specifying the boundries of each bin
plt.hist(flowers_df.sepal_width, bins=np.arange(2,5,0.25));
```

```python
# Bins of unequal sizes
plt.hist(flowers_df.sepal_width, bins=[1,3,4,4.5]);
```

# Multiple Histograms

Similar to line charts we can draw multiple histograms in a single chart.We can reduce the opacity of each histogram so the the bars of one histogram dont hide the bars for others

```python
setosa_df= flowers_df[flowers_df.species=='setosa']
versicolor_df= flowers_df[flowers_df.species=='versicolor']
virginica_df= flowers_df[flowers_df.species=='virgnica']
```

```python
plt.hist(setosa_df.sepal_width, alpha=0.4 ,bins=np.arange(2,5,0.25));
plt.hist(versicolor_df.sepal_width, alpha=0.4, bins=np.arange(2,5,0.25));
plt.legend(['Setosa','Versicolor']);
```

We can also stack multiple histograms on top of one another

```python
plt.title('Distribution of Sepal width')
plt.hist([setosa_df.sepal_width,versicolor_df.sepal_width, virginica_df.sepal_width], bins=np.arange(2,5,0.25), stacked=True);
plt.legend(['Setosa', 'Versicolor','Virginica']);
```

# Bar chart

Bar charts quite similar to line charts they show a squence of values howver a bar is show for each value rather than points connected by lines we can use the plt.bar function to draw a bar chart

```python
years=range(2000,2006)
apples=[0.35,0.6,0.9,0.8,0.65,0.8]
oranges=[0.4,0.8,0.9,0.7,0.6,0.8]
plt.bar(years, oranges);
```

```python
plt.plot(years,apples)
```

```python
plt.bar(years,oranges);
plt.plot(years, oranges, 'o--r');
plt.title('Yield of Oranges')
```

Like histograms bars can also be stacked on top of one another we use the bottom argument to plt.bar to achieve this

```python
plt.bar(years, apples)
plt.bar(years, oranges, bottom=apples);
```

# Bar Plots with Averages

Lets look at another sample dataset included with seaborn called 'tips'the dataset contains information about the sex time of day total bill and tip amount for customers visiting a restaurent over a week

```python
tips_df=sns.load_dataset('tips');
tips_df
```

We might want to draw a bar chart to visualize how the average bill amount varies across different days of the week one way to do this would be to compute the day-wise average and the use plt.bar

```python
plt.bar(tips_df.total_bill, tips_df.day);
```

However since this is a very common use  case the seaborn library provides a barplot function which can automatically compute averages

```python
sns.barplot('day', 'total_bill',  data=tips_df);
```

```python
bill_avg_df=tips_df.groupby('day')[['total_bill']].mean()
```

```python
plt.bar(bill_avg_df.index, bill_avg_df.total_bill)
```

```python
sns.barplot('day', 'total_bill', hue='sex', data=tips_df);
```

```python
sns.barplot('day', 'total_bill', hue='smoker', data=tips_df);
```

you can make the bars horizontal simply by switching the axes

```python
sns.barplot('total_bill','day', hue='smoker', data=tips_df);
```

# Heatmap

A heatmap is used to visualize 2-dimensional data like a matrix or a table using colors

```python
df=sns.load_dataset('flights')
df
```

```python
plt.plot(df.passengers);
```

```python
flights_df=sns.load_dataset('flights').pivot('month','year','passengers')
flights_df
```

flights_df is a matrix with one rpw for each month and one column of each year the values in the matrix show the number of passenger that visited te airport in a specific month of a specific year we canuse the sns.heatmap function to visualize the footfall at the airport

```python
plt.title('No. of Passengers (1000s)')
sns.heatmap(flights_df)
```

```python
plt.title('No. of Passengers (1000s)')
sns.heatmap(flights_df, fmt='d', annot=True, cmap='Blues');
```

# Plotting Multiples Charts in grid

Matplotlib and seaborn also support plotting multiple charts in a grid using plt.subplots which return a set of axes that can be used for plotting

```python
fig,axes=plt.subplots(2,3,figsize=(16,8))
axes[0,0].plot(years, apples, 's-b')
axes[0,0].plot(years,oranges, 'o--r')
axes[0,0].set_xlabel('Year')
axes[0,0].set_ylabel('Yield (tons per hectare)')
axes[0,0].set_title('Yield of oranges')
axes[0,0].legend(['Apples', 'Oranges']);
axes[0,1].set_title('Sepal lenght vs. widht')
sns.scatterplot(flowers_df.sepal_length,
               flowers_df.sepal_width,
               hue=flowers_df.species,
               s=100,
               ax=axes[0,1]);

axes[0,2].set_title('Distribution of Sepal Widht')
axes[0,2].hist([setosa_df.sepal_width, versicolor_df.sepal_width, virginica_df.sepal_width],bins=np.arange(2,5,0.25), stacked=True);

axes[1,0].set_title('Restuarant bills')
sns.barplot('day','total_bill', hue='sex',data=tips_df, ax=axes[1,0]);

axes[1,1].set_title('Flight traffic')
sns.heatmap(flights_df, cmap='Blues', ax=axes[1,1]);

axes[1,2].set_title('Steve Smith')
axes[1,2].imshow(img)
axes[1,2].grid(False)
axes[1,2].set_xticks([])
axes[1,2].set_yticks([])



plt.tight_layout(pad=2);
```
